```python
def url_link(base_url):
    start_date = datetime.date(2023, 11, 15)
    end_date = datetime.date(2024, 1, 4)   
    current_date = start_date
    while current_date <= end_date:
        formatted_date = current_date.strftime("%d/%m/%Y")
        url = base_url.format(formatted_date)
        today = datetime.date.today()
        days_left = (current_date - today).days
        urls.append((url, days_left))
        current_date += datetime.timedelta(days=1)
```

```python
driver = webdriver.Chrome()

air_lines = []
flight_number = []
source_city = []
source_time = []
destination_city = []
destination_time = []
duration = []
days_left = []
stops = []
price = []

for url in urls:
    driver.get(url[0])
    days_left.append(url[1])
    class_name1 = "airl-txt-n"
    elements = driver.find_elements(By.CLASS_NAME, class_name1)
    for element in elements:
        split_al = element.text.split('\n')
        air_lines.append(split_al[0])
        flight_number.append(split_al[1])
        
    class_name2 = "top5"
    elements = driver.find_elements(By.CLASS_NAME, class_name2)
    for i in range(0, len(elements), 2):
        split_sl = elements[i].text.split('\n')
        source_time.append(split_sl[0])
        source_city.append(split_sl[1])
    for i in range(1, len(elements), 2):
        split_sl = elements[i].text.split('\n')
        destination_time.append(split_sl[0])
        destination_city.append(split_sl[1])
    
    class_name3 = "non-st"
    elements = driver.find_elements(By.CLASS_NAME, class_name3)
    for element in elements:
        split_du = element.text.split('\n')
        duration.append(split_du[0])
        stops.append(split_du[1])
    
    class_name4 = "cle"
    elements = driver.find_elements(By.CLASS_NAME, class_name4)
    for element in elements:
        split_pr = element.text.split('\n')
        price.append(split_pr[0])
```

```python
df = pd.DataFrame({'airline': air_lines, 'flight': flight_number, 'source_city': source_city, 
                         'departure_time': source_time, 'stops': stops, 'arrival_time': destination_time, 
                        'destination_city': destination_city, 'class': 'Business', 
                   'duration': duration, 'days_left': days_left, 'price': price})
```

```python
def label_time_of_day(time_str):
    hour = int(time_str.split(':')[0])
    if 0 <= hour < 3:
        return 'Late_Night'
    elif 3 <= hour < 6:
        return 'Early_Morning'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
```

```python
df.departure_time = df.departure_time.apply(label_time_of_day)
df.arrival_time = df.arrival_time.apply(label_time_of_day)
```

```python
def stop_conversion(description):
    if 'non-stop' in description:
        return 'zero'
    elif '1-stop' in description:
        return 'one'
    else:
        return 'two_or_more'
```

```python
df.stops = df.stops.apply(stop_conversion)
```

```python
def duration_conversion(duration):
    hours, minutes = duration.split()
    hours = int(hours[:-1]) 
    minutes = int(minutes[:-1]) 
    decimal_hours = hours + (minutes / 60)
    return round(decimal_hours, 2)
```

```python
df.duration = df.duration.apply(duration_conversion)
```

```python
df.price = df.price.apply(lambda x: int(x.replace(',', '')))
```
